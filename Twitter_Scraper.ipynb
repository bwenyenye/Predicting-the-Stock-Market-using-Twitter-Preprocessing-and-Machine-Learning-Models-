**Adding a Progress Bar**

```python
scraper = sntwitter.TwitterSearchScraper("balwin properties")

tweets = []
n_tweets = 1_0000

for i, tweet in tqdm(enumerate ( scraper.get_items()), total = n_tweets):
  data = [tweet.date,
        tweet.id,
        tweet.content, 
        tweet.username,
        ]
  tweets.append(data)
  if i > n_tweets:
    break
```

```python
tweet_df = pd.DataFrame(tweets, columns = ['date', 'id', 'content', 'username'])
```

```python
tweet_df
```

```python
tweet_df.to_csv("python-tweets.csv", index = False)
```
